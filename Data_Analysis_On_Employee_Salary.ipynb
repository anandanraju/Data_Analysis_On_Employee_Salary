```python
mydata_mean = mydata.groupby('Year')['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits'].mean()
mydata_mean
```

```python
mydata_median = mydata.groupby('Year')['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits'].median()
mydata_median
```

```python
mydata_std = mydata.groupby('Year')['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits'].std()
mydata_std
```

```python
mydata_var = mydata.groupby('Year')['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits'].var()
mydata_var
```

```python
mydata_corr = mydata.groupby('Year')['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits'].corr()
mydata_corr
```
